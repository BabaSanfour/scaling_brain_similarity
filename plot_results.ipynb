```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
plots_folder = "Plots"
```

```python
if not os.path.exists(plots_folder):
    os.makedirs(plots_folder)
```

```python
data = pd.read_pickle('brain_score_results/resnet50_scaling_res.pkl')
parameter_counts = []
IT_values = []
v2_values = []
v4_values = []
v1_values = []

for key, value in data.items():
    parameter_count = int(key.split(":")[1].split()[0])
    parameter_counts.append(parameter_count)
    
    IT_values.append(value['IT'][0])
    v2_values.append(value['v2'][0])
    v4_values.append(value['v4'][0])
    v1_values.append(value['v1'][0])
Accuracy = [47.396, 54.064, 54.246, 65.276, 63.564, 68.564, 69.543]
print("Number of Parameters:", parameter_counts)
print("IT Values:", IT_values)
print("v2 Values:", v2_values)
print("v4 Values:", v4_values)
print("v1 Values:", v1_values)
print("Accuracy:", Accuracy)
```

```python
# Create figure and axis with specific size
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot IT, v2, v4, and v1 on the first y-axis (left)
ax1.loglog(parameter_counts, IT_values, marker='o', markersize=8, label='IT', color='b', linestyle='-', linewidth=2)
ax1.loglog(parameter_counts, v2_values, marker='o', markersize=8, label='v2', color='g', linestyle='-', linewidth=2)
ax1.loglog(parameter_counts, v4_values, marker='o', markersize=8, label='v4', color='r', linestyle='-', linewidth=2)
ax1.loglog(parameter_counts, v1_values, marker='o', markersize=8, label='v1', color='c', linestyle='-', linewidth=2)

# Create a second y-axis (right) and plot accuracy
ax2 = ax1.twinx()
ax2.loglog(parameter_counts, Accuracy, marker='o', markersize=8, label='Accuracy', color='m', linestyle='-', linewidth=2)

# Set labels and titles with specific font size
ax1.set_xlabel('Number of Parameters (log scale)', fontsize=14, weight = "bold")
ax1.set_ylabel('IT, v2, v4, v1 Similarity Values', color='k', fontsize=14, weight = "bold")
ax2.set_ylabel('Accuracy', color='k', fontsize=14, weight = "bold")
ax1.tick_params(axis='both', which='major', labelsize=20, width=2, length=6)
ax2.tick_params(axis='y', which='major', labelsize=20, width=2, length=6)
# Set legend with specific font size
ax1.legend(loc='center left', fontsize=14)
ax2.legend(loc='center right', fontsize=14)

# Add gridlines
ax1.grid(True, which="both", ls="--", linewidth=0.5)

# Show the plot
plt.title('ResNet50 Similarity Values and Accuracy vs. Number of Parameters', fontsize=20)
plt.tight_layout()
plt.show()
fig.savefig(os.path.join(plots_folder, 'resnet50_scaling.png'))
```

```python
data = pd.read_pickle('brain_score_results/scaling_data_imagnet.pkl')
data_counts = []
IT_values = []
v2_values = []
v4_values = []
v1_values = []

for key, value in data.items():
    data_count = int(key.split(":")[1].split()[0])
    data_counts.append(data_count)
    
    IT_values.append(value['IT'][0])
    v2_values.append(value['v2'][0])
    v4_values.append(value['v4'][0])
    v1_values.append(value['v1'][0])
Accuracy = [0.16, 0.388, 0.5, 1.256, 3.736, 8.04, 9.844, 23.016, 34.104, 51.08, 61.17, 69.14, 69.2]
IT_values.reverse()
v2_values.reverse()
v4_values.reverse()
v1_values.reverse()
Accuracy.reverse()
data_counts.reverse()
print("Number of Training samples:", data_counts)
print("IT Values:", IT_values)
print("v2 Values:", v2_values)
print("v4 Values:", v4_values)
print("v1 Values:", v1_values)
print("Accuracy:", Accuracy)

data_places = pd.read_pickle('brain_score_results/scaling_data_places365.pkl')
data_counts_places365 = []
IT_values_places365 = []
v2_values_places365 = []
v4_values_places365 = []
v1_values_places365 = []

for key, value in data_places.items():
    data_count = int(key.split(":")[1].split()[0])
    data_counts_places365.append(data_count)
    
    IT_values_places365.append(value['IT'][0])
    v2_values_places365.append(value['v2'][0])
    v4_values_places365.append(value['v4'][0])
    v1_values_places365.append(value['v1'][0])
Accuracy_places365 = [0.6356, 1.92, 2.15, 5.677, 13.075, 9.98, 23.825, 31.77, 31.271,]# 42.433, 47.93]
IT_values_places365.reverse()
v2_values_places365.reverse()
v4_values_places365.reverse()
v1_values_places365.reverse()
Accuracy_places365.reverse()
data_counts_places365.reverse()
print("Number of Training samples (Places365):", data_counts_places365)
print("IT Values (Places365):", IT_values_places365)
print("v2 Values (Places365):", v2_values_places365)
print("v4 Values (Places365):", v4_values_places365)
print("v1 Values (Places365):", v1_values_places365)
print("Accuracy (Places365):", Accuracy_places365)
```

```python
# Create figure and axis with specific size
fig, ax1 = plt.subplots(figsize=(16, 12))

# Plot IT, v2, v4, and v1 on the first y-axis (left)
ax1.loglog(data_counts, IT_values, marker='o', markersize=8, label='IT imagenet', color='b', linestyle='-', linewidth=2)
ax1.loglog(data_counts, v2_values, marker='o', markersize=8, label='v2 imagenet', color='g', linestyle='-', linewidth=2)
ax1.loglog(data_counts, v4_values, marker='o', markersize=8, label='v4 imagenet', color='r', linestyle='-', linewidth=2)
ax1.loglog(data_counts, v1_values, marker='o', markersize=8, label='v1 imagenet', color='c', linestyle='-', linewidth=2)

ax1.loglog(data_counts_places365, IT_values_places365, marker='o', markersize=8, label='IT places365', color='b', linestyle='--', linewidth=2)
ax1.loglog(data_counts_places365, v2_values_places365, marker='o', markersize=8, label='v2 places365', color='g', linestyle='--', linewidth=2)
ax1.loglog(data_counts_places365, v4_values_places365, marker='o', markersize=8, label='v4 places365', color='r', linestyle='--', linewidth=2)
ax1.loglog(data_counts_places365, v1_values_places365, marker='o', markersize=8, label='v1 places365', color='c', linestyle='--', linewidth=2)

# Create a second y-axis (right) and plot accuracy
ax2 = ax1.twinx()
ax2.loglog(data_counts, Accuracy, marker='o', markersize=8, label='Accuracy ImageNet', color='m', linestyle='-', linewidth=2)
ax2.loglog(data_counts_places365, Accuracy_places365, marker='o', markersize=8, label='Accuracy Places365', color='m', linestyle='--', linewidth=2)

# Set labels and titles with specific font size
ax1.set_xlabel('Training Data Size (log scale)', fontsize=14, weight = "bold")
ax1.set_ylabel('IT, v2, v4, v1 Similarity Values', color='k', fontsize=14, weight = "bold")
ax2.set_ylabel('Accuracy', color='k', fontsize=14, weight = "bold")
ax1.tick_params(axis='both', which='major', labelsize=20, width=2, length=6)
ax2.tick_params(axis='y', which='major', labelsize=20, width=2, length=6)
# Set legend with specific font size
ax1.legend(loc='upper left', fontsize=14)
ax2.legend(loc='lower right', fontsize=14)

# Add gridlines
ax1.grid(True, which="both", ls="--", linewidth=0.5)

# Show the plot
plt.title('Imagenet: Similarity Values and Accuracy vs. Number of Training Samples', fontsize=20)
plt.tight_layout()
plt.show()
fig.savefig(os.path.join(plots_folder, 'data_scaling.png'))
```

```python
data = pd.read_pickle('brain_score_results/compute_results.pkl')
epochs_counts = []
IT_values = []
v2_values = []
v4_values = []
v1_values = []
Accuracy = []
for key, value in data.items():
    epochs_count = int(key.split(":")[1])
    epochs_counts.append(epochs_count)
    
    IT_values.append(value['IT'][0])
    v2_values.append(value['v2'][0])
    v4_values.append(value['v4'][0])
    v1_values.append(value['v1'][0])
    Accuracy.append(value['acc'])
print("Number of Epochs:", epochs_counts)
print("IT Values:", IT_values)
print("v2 Values:", v2_values)
print("v4 Values:", v4_values)
print("v1 Values:", v1_values)
print("Accuracy:", Accuracy)
```

```python
# Create figure and axis with specific size
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot IT, v2, v4, and v1 on the first y-axis (left)
ax1.loglog(epochs_counts, IT_values, marker='o', markersize=8, label='IT', color='b', linestyle='-', linewidth=2)
ax1.loglog(epochs_counts, v2_values, marker='o', markersize=8, label='v2', color='g', linestyle='-', linewidth=2)
ax1.loglog(epochs_counts, v4_values, marker='o', markersize=8, label='v4', color='r', linestyle='-', linewidth=2)
ax1.loglog(epochs_counts, v1_values, marker='o', markersize=8, label='v1', color='c', linestyle='-', linewidth=2)

# Create a second y-axis (right) and plot accuracy
ax2 = ax1.twinx()
ax2.loglog(epochs_counts, Accuracy, marker='o', markersize=8, label='Accuracy', color='m', linestyle='-', linewidth=2)

# Set labels and titles with specific font size
ax1.set_xlabel('Number of Parameters (log scale)', fontsize=14, weight = "bold")
ax1.set_ylabel('IT, v2, v4, v1 Similarity Values', color='k', fontsize=14, weight = "bold")
ax2.set_ylabel('Accuracy', color='k', fontsize=14, weight = "bold")
ax1.tick_params(axis='both', which='major', labelsize=20, width=2, length=6)
ax2.tick_params(axis='y', which='major', labelsize=20, width=2, length=6)
# Set legend with specific font size
ax1.legend(loc='center left', fontsize=14)
ax2.legend(loc='center right', fontsize=14)

# Add gridlines
ax1.grid(True, which="both", ls="--", linewidth=0.5)

# Show the plot
plt.title('Similarity Values and Accuracy vs. Number of Epochs', fontsize=20)
plt.tight_layout()
plt.show()
fig.savefig(os.path.join(plots_folder, 'compute_scaling.png'))
```
